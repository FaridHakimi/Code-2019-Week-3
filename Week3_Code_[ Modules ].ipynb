## Timing difference

```python
import timeit as time

setup="import numpy as np;import math;a=(100.5,200.3,500.3,55.345,22.3);b=np.array(a)"

numpy = time.Timer("np.floor(b)",setup=setup)
math = time.Timer("[math.floor(i) for i in a]",setup=setup)

print(numpy.timeit())
print(math.timeit())
```

## Random numbers

```python
"""
We can pick random numbers using numpy
"""
import numpy.random as random

print(random.randint(10)) #picks a number between 0 and 10

# Try running this cell several times, the number will change
```

```python
"""
There is a significant difference between numpy and random in the randomness
"""

import numpy.random as nprnd
import numpy as np
import random as rnd
import matplotlib.pyplot as plt

start, stop = 1,10
size = (10,100,1000)

numpya,randomA = [],[]
npb, randb = [],[]
npc, randomc = [],[]


"""

Generating enormous amounts of random data to fit matplotplib.

"""

for i in size:
    numpya.append(nprnd.randint(start,stop,i))
    
    out = []
    for n in range(i):
        out.append(rnd.randrange(start,stop))
    randoma.append(out)
    
for p in range(len(size)):
    npb.append([np.where(np.array(numpya[p])==i)[0] for i in range(start,stop)])
    randb.append([np.where(np.array(randoma[p])==i)[0] for i in range(start,stop)])

for f in range(len(size)):
    npc.append([len(n)/size[f] for n in npb[f]])
    
for f in range(len(size)):
    randomc.append([len(n)/size[f] for n in randb[f]])
    
```

```python

plt.plot(npc[0],label="Numpy")
plt.plot(randomc[0],label="Random")
plt.legend()    
plt.show()

```

```python
plt.plot(npc[1],label="Numpy")
plt.plot(randomc[1],label="Random")
plt.legend()    
plt.show()
```

```python
plt.plot(npc[2],label="Numpy")
plt.plot(randomc[2],label="Random")
plt.legend()    
plt.show()
```
